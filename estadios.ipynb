```python
center_lat = -14.24007
center_lon = -53.18050
```

```python
consulta = list(estadios.find())
```

```python
df_convertido = pd.DataFrame(consulta)
```

```python
df_convertido.iloc[0]
```

```python
locationmap = folium.Map(location=[center_lat, center_lon], zoom_start=5, width=800, height=600)
```

```python
qtde = len(df_convertido)
```

```python
qtde
```

```python
for i in range(qtde):
    lng = float(df_convertido.iloc[i]['longitude'])
    lat = float(df_convertido.iloc[i]['latitude'])
    nome = df_convertido.iloc[i]['nome']
    capacidade = df_convertido.iloc[i]['capacidade']

    description = f'Estádio: {nome}<br>Capacidade: {capacidade}'

    folium.Marker(
        location=[lat, lng],
        popup=description,
        icon=folium.Icon(color='red')
    ).add_to(locationmap)
```
